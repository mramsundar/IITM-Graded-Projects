```python
import pandas as pd
import numpy as np
from pandasql import sqldf
import matplotlib.pyplot as plt
from IPython.display import Markdown, display

# load the dataset
cust_df = pd.read_csv("Customer Satisfaction.csv")

display(Markdown("# 1. Gender Distribution"))

gender_counts = cust_df["Gender"].value_counts()
gender_proportion = gender_counts / len(cust_df) * 100

# Create a figure with two subplots side by side
fig, (left_chart, right_chart) = plt.subplots(1, 2, figsize=(12, 5))

# First subplot - pie chart
left_chart.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", startangle=45)
left_chart.set_title("Male to Female travellers proportion")

# SQL query to find out satisfaction counts by the Gender
query = """
select distinct 
    Gender, Satisfaction, count(distinct id) as DistinctCustomers
from 
    cust_df
group by
    Gender, Satisfaction
"""

satisfaction_result_df = sqldf(query, globals())

# Create pivot table for easier plotting
satisfaction_pivot_df = satisfaction_result_df.pivot(index="Gender", columns="Satisfaction", values="DistinctCustomers")

# Second subplot - bar chart
satisfaction_pivot_df.plot(kind="bar", ax=right_chart)
right_chart.set_title("Gender v/s Satisfaction Levels")
right_chart.set_xlabel("Gender")
right_chart.set_ylabel("# Customers")
right_chart.legend(title="Satisfaction Level")
right_chart.tick_params(axis="x", rotation=0)

# Add value labels
for container in right_chart.containers:
    right_chart.bar_label(container, label_type="edge")


plt.show()



display(Markdown("# 2. Age Analysis"))
"""
What is the average age of travellers in the dataset? 
What is the age range of the travellers, and how many travellers fall within specific age brackets (e.g., 18-30, 31-45, 46-60, 61+)? 
"""

query = """
select 
    avg(Age) as AverageAge
from 
    cust_df
"""

age_result_df = sqldf(query, globals())
display(Markdown(f"## The average age of a traveller is : {age_result_df.values[0][0]:.2f}" ))


bins = [0, 17, 30, 45, 60, float("inf")]
labels = ["Under 18", "18-30", "31-45", "46-60", "61+"]

cust_df_copy = cust_df.copy(deep= True)

cust_df_copy["Age Bracket"] = pd.cut(cust_df_copy["Age"], bins= bins, labels= labels, right= True, include_lowest= True)
age_bracket_counts_df = cust_df_copy["Age Bracket"].value_counts().sort_index()

# display(age_bracket_counts_df)

age_bracket_counts_df.plot(kind="bar")
plt.xlabel("Age Bracket")
plt.ylabel("Number of Travellers")
plt.title("Distribution of Travellers by Age Bracket")

for index, count in enumerate(age_bracket_counts_df):
    plt.text(index, count + 0.05, str(count), ha= "center", va= "bottom")

plt.show()





display(Markdown("# 3. Travel Category"))
"""
How many travellers are classified under "Business travel" versus "Personal Travel"? 

What is the average distance travelled by each travel category? 
"""


query = """
select 'Business Travel' as [Travel Type],   count(distinct id) as [# Travellers], avg([Distance Travelled]) as [Avg. Distance Travelled] from cust_df where [Travel Category] = 'Official'
union all
select  'Personal Travel', count(distinct id), avg([Distance Travelled]) from cust_df where [Travel Category] = 'Personal Travel'
"""

display(sqldf(query, globals()))




display(Markdown("# 4. Travel Class Ratings"))
"""
What is the average rating for seat comfort across different travel classes? 

How does the rating for food differ between "Business" and "Eco" classes? 

"""

query = """

select 
    coalesce([Travel Class], 'Not Specified') as [Travel Class]
    , avg([Seat Comfort]) as [Average Rating for Seat Comfort]
from
    cust_df
group by 
    [Travel Class]

"""

display(Markdown("## Average rating for seat comfort across travel classes"))
seat_comfort_df = sqldf(query, globals())
# display(seat_comfort_df)


plt.figure(figsize=(8, 4))
bars = plt.bar(seat_comfort_df["Travel Class"], seat_comfort_df["Average Rating for Seat Comfort"], color='skyblue')
plt.xlabel("Travel Class")
plt.ylabel("Average Rating for Seat Comfort")
plt.title("Average Seat Comfort Rating by Travel Class")
plt.ylim(0, 5)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f"{yval:.2f}", ha='center', va='bottom')

plt.show()


query = """

select 'Business' as [Travel Class], avg([Food]) as [Average Food Rating] from cust_df where [Travel Class] = 'Business'
union all
select 'Economy' as [Travel Class], avg([Food]) as [Average Food Rating] from cust_df where [Travel Class] = 'Economy'
"""
display(Markdown("## Rating for food between Business and Eco class"))
display(sqldf(query, globals()))



display(Markdown("# 5. Delay Analysis"))
"""
What is the average departure delay and arrival delay for the dataset? 
How do delays (both departure and arrival) impact customer satisfaction? 

"""

query = """
select avg([Departure Delay (min)]) as [Average Departure Delay], avg([Arrival Delay (min)]) as [Average Arrival Delay] from cust_df
"""
display(Markdown("## Average departure and arrival delays in the dataset (in minutes)"))
display(sqldf(query, globals()))

display(Markdown("## How delayed departure time affects customer satisfaction"))

# just get necessary columns only
cols_needed = ["Departure Delay (min)", "Arrival Delay (min)", "Satisfaction"]
delay_df = (
    cust_df
      .loc[:, cols_needed]                          
      .dropna(subset=cols_needed)
      .assign(total_delay=lambda d: d["Departure Delay (min)"] + d["Arrival Delay (min)"])
      .copy()
)


#create bins for time categorization
bins   = [-0.1, 0, 5, 15, 30, 60, np.inf]
labels = ["0", "1-5", "6-15", "16-30", "31-60", ">60"]

delay_df["Overall Delay Range (In Minutes)"] = pd.cut(
    delay_df["total_delay"],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

bucket_counts = (
    pd.crosstab(delay_df["Overall Delay Range (In Minutes)"], delay_df["Satisfaction"])
      .reindex(labels, fill_value=0)
)
bucket_counts["Total"] = bucket_counts.sum(axis=1)

display(bucket_counts)


```
